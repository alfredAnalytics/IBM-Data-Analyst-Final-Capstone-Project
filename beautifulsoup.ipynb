```python
from bs4 import BeautifulSoup

def find_title_tags(html_content):
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all title tags
    title_tags = soup.find_all('title')

    # Extract and print the contents of title tags
    for title_tag in title_tags:
        print(title_tag.string)  # or print(title_tag.text) for older versions of BeautifulSoup

# Example usage
if __name__ == "__main__":
    # Read the HTML file content
    with open('/content/BeautifulSoup.html', 'r', encoding='utf-8') as f:
        html_content = f.read()

    # Call the function to find title tags
    find_title_tags(html_content)
```

```python
#Q2 Write a Python program to get the number of paragraph tags of a given html document.
def count_paragraph_tags(html_content):
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all paragraph tags
    paragraph_tags = soup.find_all('p')

    # Count the number of paragraph tags
    num_paragraphs = len(paragraph_tags)

    return num_paragraphs

# Example usage
if __name__ == "__main__":
    # Read the HTML file content
    with open('/content/BeautifulSoup.html', 'r', encoding='utf-8') as f:
        html_content = f.read()

    # Call the function to count paragraph tags
    num_paragraphs = count_paragraph_tags(html_content)

    print(f"Number of <p> tags: {num_paragraphs}")
```

```python
#Q3.	Write a Python program to extract the text in the first paragraph tag of a given html document.
```

```python
def extract_first_paragraph_text(html_content):
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find the first paragraph tag
    first_paragraph = soup.find('p')

    # Extract and return the text content if found
    if first_paragraph:
        return first_paragraph.text
    else:
        return None

# Example usage
if __name__ == "__main__":
    # Read the HTML file content
    with open('/content/BeautifulSoup.html', 'r', encoding='utf-8') as f:
        html_content = f.read()

    # Call the function to extract the text
    text = extract_first_paragraph_text(html_content)

    if text:
        print(text)
    else:
        print("No paragraph tag found.")
```

```python
#Q4 Write a Python program to extract all the URLs from the webpage python.org that are nested within <li> tags form
```

```python
import requests

def extract_urls_from_li(url):
    response = requests.get(url)
    response.raise_for_status()  # Check for HTTP errors

    soup = BeautifulSoup(response.content, 'html.parser')

    urls = []
    for li in soup.find_all('li'):
        for a in li.find_all('a', href=True):
            urls.append(a['href'])

    return urls

if __name__ == "__main__":
    url = "https://www.python.org/"
    extracted_urls = extract_urls_from_li(url)

    for url in extracted_urls:
        print(url)
```

```python
#Q5.	Write a Python program to extract all the text from a given web page.
```

```python
import requests
from bs4 import BeautifulSoup

def extract_text_from_webpage(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        # Extract all text from the parsed HTML
        text = soup.get_text()

        return text
    else:
        print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
        return None

# Example usage
if __name__ == "__main__":
    # Example URL
    url = 'https://en.wikipedia.org/wiki/Python_(programming_language)'

    # Call the function to extract text from the webpage
    webpage_text = extract_text_from_webpage(url)

    if webpage_text:
        # Print the extracted text (you can also save it to a file, process it further, etc.)
        print(webpage_text)
```

```python
import requests
from bs4 import BeautifulSoup

def retrieve_title_info(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find the title element
        title_tag = soup.title

        if title_tag:
            # Retrieve the HTML code of the title element
            title_html = str(title_tag)

            # Retrieve the text content of the title element
            title_text = title_tag.get_text()

            # Retrieve the HTML code of the parent element of the title
            parent_html = str(title_tag.parent)

            return title_html, title_text, parent_html
        else:
            print("No title element found in the webpage.")
            return None
    else:
        print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
        return None

# Example usage
if __name__ == "__main__":
    # Example URL
    url = 'https://en.wikipedia.org/wiki/Python_(programming_language)'

    # Call the function to retrieve title information
    title_html, title_text, parent_html = retrieve_title_info(url)

    if title_html and title_text and parent_html:
        # Print the retrieved information
        print(f"HTML code of the title:\n{title_html}\n")
        print(f"Text content of the title:\n{title_text}\n")
        print(f"HTML code of the parent of the title:\n{parent_html}\n")
```

```python
import requests
from bs4 import BeautifulSoup

def format_html_to_unicode(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        # Prettify the parse tree into a nicely formatted Unicode string
        pretty_html = soup.prettify()

        return pretty_html
    else:
        print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
        return None

# Example usage
if __name__ == "__main__":
    # Example URL
    url = 'https://en.wikipedia.org/wiki/Python_(programming_language)'

    # Call the function to format HTML to Unicode
    formatted_html = format_html_to_unicode(url)

    if formatted_html:
        # Print the formatted HTML
        print(formatted_html)
```

```python
#Q8 Write a Python program to remove the contents of a tag in a given html document.
```

```python
from bs4 import BeautifulSoup

def remove_tag_contents(html_content, tag_name):
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all instances of the specified tag
    tags = soup.find_all(tag_name)

    # Remove the contents of each tag found
    for tag in tags:
        tag.clear()  # Clear the contents of the tag

    # Return the modified HTML content
    return str(soup)

# Example usage
if __name__ == "__main__":
    # Example HTML content
    html_content = """
    <!DOCTYPE html>
    <html>
    <head>
        <title>Example HTML Document</title>
    </head>
    <body>
        <h1>This is a Heading</h1>
        <p>This is a paragraph with <strong>bold</strong> text.</p>
        <div>
            <p>This is another paragraph in a <span>div</span>.</p>
        </div>
    </body>
    </html>
    """

    # Specify the tag name whose contents you want to remove
    tag_name = 'p'

    # Call the function to remove tag contents
    modified_html = remove_tag_contents(html_content, tag_name)

    # Print the modified HTML
    print(modified_html)
```

```python
#Q9 Write a Python program to wrap an element in the specified tag and create the new wrapper.
```

```python
from bs4 import BeautifulSoup

def wrap_element_in_tag(html_content, tag_name, new_tag):
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find the element to be wrapped
    element = soup.find(tag_name)

    if element:
        # Create a new tag
        wrapper_tag = soup.new_tag(new_tag)

        # Move the element inside the new tag
        element.wrap(wrapper_tag)

        # Return the modified HTML content
        return str(soup)
    else:
        print(f"Element with tag '{tag_name}' not found.")
        return None

# Example usage
if __name__ == "__main__":
    # Example HTML content
    html_content = """
    <!DOCTYPE html>
    <html>
    <head>
        <title>Example HTML Document</title>
    </head>
    <body>
        <h1>This is a Heading</h1>
        <p>This is a paragraph.</p>
        <div>
            <p>This is another paragraph in a <span>div</span>.</p>
        </div>
    </body>
    </html>
    """

    # Specifying the tag name of the element to wrap and the new tag name
    tag_name = 'p'
    new_tag = 'div'  # New tag to wrap the 'p' element

    # Call the function to wrap the element
    modified_html = wrap_element_in_tag(html_content, tag_name, new_tag)

    # Print the modified HTML
    print(modified_html)
```

```python
from bs4 import BeautifulSoup

def remove_tag_and_contents(html_content, tag_name):
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all instances of the specified tag
    tags = soup.find_all(tag_name)

    # Remove the tag and its contents
    for tag in tags:
        tag.decompose()  # Destroy the tag and its contents

    # Return the modified HTML content
    return str(soup)

# Example usage
if __name__ == "__main__":
    # Example HTML content
    html_content = """
    <!DOCTYPE html>
    <html>
    <head>
        <title>Example HTML Document</title>
    </head>
    <body>
        <h1>This is a Heading</h1>
        <p>This is a paragraph.</p>
        <div>
            <p>This is another paragraph in a <span>div</span>.</p>
        </div>
    </body>
    </html>
    """

    # Specifying the tag name to remove
    tag_name = 'div'

    # Call the function to remove the tag and its contents
    modified_html = remove_tag_and_contents(html_content, tag_name)

    # Print the modified HTML
    print(modified_html)
```
